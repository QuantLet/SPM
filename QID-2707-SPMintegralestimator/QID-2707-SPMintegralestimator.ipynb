```python
import numpy as np
import matplotlib.pyplot as plt

# Define the logistic function
def logistic(t, alpha=0):
    return 1 / (1 + np.exp(-t + alpha))

# Generate the values
t = np.linspace(-6, 9, 400)
F_t = logistic(t)           # First logistic curve F(t)
F_t_alpha = logistic(t, alpha=3)  # Second logistic curve F(t+alpha)

# Create the plot
plt.figure(figsize=(8,6))

# Plot both logistic functions in black
plt.plot(t, F_t, color="black")
plt.plot(t, F_t_alpha, color="black")

# Fill the area between the curves with yellow color
plt.fill_between(t, F_t, F_t_alpha, color="yellow", alpha=0.7)

# Labeling the distance β and positioning it in the middle of the shaded area
beta_x = (t[len(t) // 2])  # Middle of the x range
beta_y = (F_t[len(t) // 2] + F_t_alpha[len(t) // 2]) / 2  # Mid-point between the two curves at that point
plt.annotate(r'$\beta$', xy=(beta_x, beta_y), fontsize=20, ha='center', va='center')

# Formatting the plot
plt.title("Integral", fontsize=16)
plt.xlabel("Index Values t", fontsize=14)
plt.ylabel("Link Functions F(t), F(t+α)", fontsize=14)

# Show plot
plt.show()
```
