```python
import matplotlib.pyplot as plt
import numpy as np

# Function to read stock data from a file
def read_stock_data(filename):
    with open(filename, 'r') as file:
        data = np.loadtxt(file)
    return data

# Read the stock data from the 'stockres.dat' file
filename = 'stockres.dat'
stock_data = read_stock_data(filename)

# Define bin widths
bin_widths = [0.007, 0.05, 0.02, 0.1]

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot histograms with different bin widths
for ax, h in zip(axs.ravel(), bin_widths):
    bins = np.arange(min(stock_data), max(stock_data) + h, h)
    ax.hist(stock_data, bins=bins, edgecolor='black', color='white')
    ax.set_title(f'h={h}, x0=0')
    ax.grid(True, which='both', linestyle='-', linewidth=0.5)
    
    # Remove the lines inside the plots
    ax.grid(False)
    ax.set_axisbelow(True)
    
    # Customize the appearance of the plot
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_linewidth(0.5)
    ax.spines['bottom'].set_linewidth(0.5)

# Adjust layout
plt.tight_layout()
plt.show()
```
