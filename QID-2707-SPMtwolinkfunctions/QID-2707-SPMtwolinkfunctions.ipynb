```python
import numpy as np
import matplotlib.pyplot as plt

# Define the index (X-axis values)
x = np.linspace(-3, 4, 400)

# Define the two link functions (Y-axis values)
g1 = 1 / (1 + np.exp(-x))  # Logistic function
g2 = 1 - g1  # Complementary of the logistic function

# Plot the two functions
plt.figure(figsize=(6, 4))
plt.plot(x, g1, 'b', label='E(Y|X)=g(Index)', linewidth=3)
plt.plot(x, g2, 'r', label='E(Y|X)=~g(Index)', linewidth=3)

# Add vertical and horizontal dotted lines to match the plot
plt.vlines([-1, 0, 1], ymin=0, ymax=1, linestyles='dotted', colors='black')
plt.hlines(0.5, xmin=-1, xmax=1, linestyles='dotted', colors='black')

# Labels and title
plt.title('Two Link Functions')
plt.xlabel('Index')
plt.ylabel('E(Y|X)=g(Index) and E(Y|X)=~g(Index)')

# Show the plot without grid
plt.grid(False)
plt.show()
```
