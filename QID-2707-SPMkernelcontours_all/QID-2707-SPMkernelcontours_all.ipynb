```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import inv, cholesky
from scipy.special import gamma

# Product Epanechnikov
def product_Epa(u):
    k = 0.75**2 * (1 - u[:, 0]**2) * (1 - u[:, 1]**2)
    k = k * (np.abs(u[:, 0]) <= 1) * (np.abs(u[:, 1]) <= 1)
    return k

# Spherical Epanechnikov
def radial_Epa(u):
    u = np.array(u)
    d = u.shape[1]
    p = 2
    q = 1
    r = 1 - np.sqrt(np.sum(u**2, axis=1))**p
    c = p * gamma(d / 2) * gamma(1 + q + d / p) / (2 * np.pi**(d / 2) * gamma(1 + q) * gamma(d / p))
    k = c * r**q * (r >= 0)
    return k

# Choose matrix H
H = np.diag([1, 1])  # Example: Diagonal matrix

# Plot kernels
u1 = u2 = np.linspace(-1, 1, 30)
invH = inv(cholesky(H))
grid = (invH @ np.array(np.meshgrid(u1, u2)).reshape(2, -1)).T

P_Kernel = product_Epa(grid) / np.linalg.det(H)
P_Kernel = P_Kernel.reshape(len(u1), len(u2))

R_Kernel = radial_Epa(grid) / np.linalg.det(H)
R_Kernel = R_Kernel.reshape(len(u1), len(u2))

levels = np.linspace(0.1, 1, 10)

# Plot Product Kernel with transparent background
plt.figure(facecolor='none')  # Transparent figure background
plt.contour(u1, u2, P_Kernel, levels=levels, colors='blue', linewidths=2.5)  # Increase linewidth for intensity
plt.title("Product Kernel")
plt.xlabel("x1")
plt.ylabel("x2")
plt.gca().set_facecolor('none')  # Transparent plot background
plt.show()

# Plot Radial-symmetric Kernel with transparent background
plt.figure(facecolor='none')  # Transparent figure background
plt.contour(u1, u2, R_Kernel, levels=levels, colors='blue', linewidths=2.5)  # Increase linewidth for intensity
plt.title("Radial-symmetric Kernel")
plt.xlabel("x1")
plt.ylabel("x2")
plt.gca().set_facecolor('none')  # Transparent plot background
plt.show()
```
